# Renaming columns by replacing the columns name

```python
personality_scores.columns=personality_scores.columns.str.replace('Section 5 of 6','')
personality_scores.columns=personality_scores.columns.str.replace('[','')
personality_scores.columns=personality_scores.columns.str.replace(']','')

personality_scores
```

# Creating new columns of personality

```python
catogories = {
        1: {
            "trait": "Extraversion",
            "key": 1
           },
        2: {
            "trait":"Agreeableness",
            "key": 2
            },
        3: {
            "trait":"Conscientiousness",
            "key": 3
        },
        4: {
            "trait":"Neuroticism",
            "key": 4
        },
        5: {
            "trait":"Intellect",
            "key": 5
            }
        }

def sum_personality(item, key, data):
    count = 0
    for column in data.columns:
        score = item[column]
        if type(score) == str:
            result = int(score[4])
            if int(score[1]) == catogories[key]["key"]:
                count += result
    return count
```

```python
for cat in catogories:

    personality_scores[catogories[cat]['trait']] = personality_scores.apply(lambda row: sum_personality(row, cat, personality_scores), axis=1)
```

```python
#Dropping columns with all null values
personality_scores.dropna(axis = 1, inplace = True)
personality_scores.head(5)
```
