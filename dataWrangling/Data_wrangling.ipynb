# Dropping columns with all null values

```python
personality_scores.dropna(axis = 1, inplace = True)
personality_scores.head(5)
```

```python
personality_scores.rename(columns=lambda x: x.replace('Section 5 of 6 [', ''), inplace=True)
personality_scores.rename(columns=lambda x: x.replace('.]',''), inplace=True)
personality_new.rename(columns=lambda x: x.replace(']',''), inplace=True)
personality_new=personality_new.dropna(axis = 1)
personality_new.head(3)
personality_scores.head(5)
```

# Creating new columns

```python
def score(row, number, dataframe):
    '''Returns a count with
    row = 'row interations
    number = 'Personality it belongs to'
    dataframe = "new dataframe"
    '''
    count = 0
    for column in dataframe.columns:
        breakdown = row[column]
        if type(breakdown) == str:
            tup_break = breakdown[1]
            result = int(breakdown[4])
            if tup_break == number:
                count += result
    return count

#Adding columns the dataframe and applying the score function
personality_new['Extraversion'] = personality_scores.apply(lambda row : score(row,'1', personality_new), axis = "columns")
personality_new['Agreeableness'] = personality_scores.apply(lambda row : score(row,'2', personality_new), axis = "columns")
personality_new['Conscientiousness'] = personality_scores.apply(lambda row : score(row,'3', personality_new), axis = "columns")
personality_new['Emotional stability'] = personality_scores.apply(lambda row : score(row,'4', personality_new), axis = "columns")
personality_new['Intellect'] = personality_scores.apply(lambda row : score(row,'5', personality_new), axis = "columns")

display(personality_scores)
```
